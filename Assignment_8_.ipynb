# Q. Statistical Mechanics of the Quantum Oscillator

The energy eigenvalues are $E_j = \hbar\omega\!\left(j+\tfrac12\right)$, $j=0,1,2,\ldots$

---
## Part 1 — The Partition Function

$$Z(T)=\sum_{j=0}^\infty e^{-E_j/(k_BT)}
=\sum_{j=0}^\infty e^{-\hbar\omega(j+1/2)/(k_BT)}.$$

Factor out the zero-point contribution:

$$Z(T)=e^{-\hbar\omega/(2k_BT)}\sum_{j=0}^\infty \left(e^{-\hbar\omega/(k_BT)}\right)^j.$$

Let $r = e^{-\hbar\omega/(k_BT)}$.  Since $\hbar\omega>0$ and $T>0$ we have $0<r<1$, so the geometric series converges:

$$\sum_{j=0}^\infty r^j = \frac{1}{1-r}.$$

Therefore

$$\boxed{Z(T)=\frac{e^{-\hbar\omega/(2k_BT)}}{1-e^{-\hbar\omega/(k_BT)}}.}$$

---
## Part 2 — Thermodynamic Potentials

Define $\beta = 1/(k_BT)$.  Then $Z = e^{-\beta\hbar\omega/2}/(1-e^{-\beta\hbar\omega})$, so

$$\ln Z = -\frac{\beta\hbar\omega}{2} - \ln\!\left(1-e^{-\beta\hbar\omega}\right).$$

### Helmholtz Free Energy

$$F(T)=-k_BT\ln Z = k_BT\!\left[\frac{\beta\hbar\omega}{2}+\ln\!\left(1-e^{-\beta\hbar\omega}\right)\right].$$

$$\boxed{F(T)=\frac{\hbar\omega}{2}+k_BT\ln\!\left(1-e^{-\hbar\omega/(k_BT)}\right).}$$

### Internal Energy

$$U = -\frac{\partial\ln Z}{\partial\beta}.$$

Differentiate $\ln Z$:

$$\frac{\partial\ln Z}{\partial\beta}=-\frac{\hbar\omega}{2}-\frac{\partial}{\partial\beta}\ln\!\left(1-e^{-\beta\hbar\omega}\right)
=-\frac{\hbar\omega}{2}+\frac{\hbar\omega\,e^{-\beta\hbar\omega}}{1-e^{-\beta\hbar\omega}}.$$

Hence

$$U=-\frac{\partial\ln Z}{\partial\beta}=\frac{\hbar\omega}{2}
-\frac{\hbar\omega\,e^{-\beta\hbar\omega}}{1-e^{-\beta\hbar\omega}}.$$

Multiply numerator and denominator by $e^{\beta\hbar\omega}$:

$$\boxed{U(T)=\frac{\hbar\omega}{2}+\frac{\hbar\omega}{e^{\hbar\omega/(k_BT)}-1}.}$$

This is the zero-point energy plus the mean occupation number $\langle n\rangle$ times $\hbar\omega$ (Planck–Einstein formula).

---
## Part 3 — Entropy

$$S=-\left(\frac{\partial F}{\partial T}\right)_V.$$

Differentiate $F = \tfrac{\hbar\omega}{2}+k_BT\ln(1-e^{-\hbar\omega/(k_BT)})$. Let $x=\hbar\omega/(k_BT)$, so $\partial x/\partial T = -x/T$.

$$\frac{\partial F}{\partial T}=k_B\ln(1-e^{-x})+k_BT\cdot\frac{e^{-x}\cdot(x/T)}{1-e^{-x}}
=k_B\ln(1-e^{-x})+\frac{k_Bx\,e^{-x}}{1-e^{-x}}.$$

Therefore

$$S=-\frac{\partial F}{\partial T}=k_B\left[-\ln(1-e^{-x})+\frac{x\,e^{-x}}{1-e^{-x}}\right], \quad x=\frac{\hbar\omega}{k_BT}.$$

**Verification $U = F + TS$:**

$$F+TS = \frac{\hbar\omega}{2}+k_BT\ln(1-e^{-x})
+T\cdot k_B\!\left[-\ln(1-e^{-x})+\frac{x\,e^{-x}}{1-e^{-x}}\right]$$
$$=\frac{\hbar\omega}{2}+\frac{k_BT\,x\,e^{-x}}{1-e^{-x}}
=\frac{\hbar\omega}{2}+\frac{\hbar\omega\,e^{-x}}{1-e^{-x}}=U.\quad\checkmark$$

---
## Part 4 — Heat Capacity and Hyperbolic Form

### Direct differentiation

$$C_V=\frac{\partial U}{\partial T}=\frac{\partial}{\partial T}\left[\frac{\hbar\omega}{e^x-1}\right],\quad x=\frac{\hbar\omega}{k_BT}.$$

$$\frac{\partial}{\partial T}\frac{1}{e^x-1}=-\frac{e^x}{(e^x-1)^2}\cdot\frac{\partial x}{\partial T}
=\frac{e^x}{(e^x-1)^2}\cdot\frac{x}{T}.$$

So

$$C_V=k_B\frac{x^2 e^x}{(e^x-1)^2},\quad x=\frac{\hbar\omega}{k_BT}.$$

### Rewriting using $\sinh$

We use $\sinh(u)=\tfrac{e^u-e^{-u}}{2}$, so $\sinh(x/2)=\tfrac{e^{x/2}-e^{-x/2}}{2}$, hence

$$2\sinh\!\left(\frac x2\right)=e^{x/2}-e^{-x/2}\implies 4\sinh^2\!\left(\frac x2\right)=(e^{x/2}-e^{-x/2})^2.$$

Now observe:

$$(e^x-1)^2 = e^x(e^{x/2}-e^{-x/2})^2 = 4e^x\sinh^2\!\left(\frac x2\right).$$

Substituting:

$$C_V=k_B\frac{x^2 e^x}{4e^x\sinh^2(x/2)}=k_B\frac{x^2}{4\sinh^2(x/2)}.$$

With $x/2=\hbar\omega/(2k_BT)$:

$$\boxed{C_V=k_B\left(\frac{\hbar\omega}{2k_BT}\right)^2\frac{1}{\sinh^2\!\left(\dfrac{\hbar\omega}{2k_BT}\right)}.}$$

---
## Part 5 — Asymptotic Behavior (High-Temperature Limit)

In the limit $k_BT\gg\hbar\omega$, we have $x=\hbar\omega/(k_BT)\ll1$.  Using $\sinh(u)\approx u$ for small $u$:

$$C_V\approx k_B\frac{(x/2)^2}{(x/2)^2}=k_B.$$

**Yes, this matches the classical equipartition theorem:** a 1D harmonic oscillator has two quadratic degrees of freedom (kinetic $+$ potential), each contributing $\tfrac12k_B$, giving $C_V=k_B$. $\checkmark$


```python
import numpy as np
import matplotlib.pyplot as plt

# ---- Quantum oscillator: numerical plots ----
kB = 1.0    # natural units: hbar*omega = 1
hbar_omega = 1.0

T_vals = np.linspace(0.05, 5.0, 500)
x = hbar_omega / (kB * T_vals)

Z  = np.exp(-x/2) / (1 - np.exp(-x))
F  = hbar_omega/2 + kB*T_vals * np.log(1 - np.exp(-x))
U  = hbar_omega/2 + hbar_omega / (np.exp(x) - 1)
S  = kB * (-np.log(1 - np.exp(-x)) + x*np.exp(-x)/(1 - np.exp(-x)))
Cv = kB * x**2 * np.exp(x) / (np.exp(x) - 1)**2

fig, axes = plt.subplots(2, 2, figsize=(11, 8))
fig.suptitle(r"Quantum Harmonic Oscillator ($\hbar\omega = k_B = 1$)", fontsize=14)

axes[0,0].plot(T_vals, U, 'b'); axes[0,0].set_title("Internal Energy U(T)")
axes[0,0].set_xlabel("T"); axes[0,0].set_ylabel("U"); axes[0,0].grid(alpha=.3)

axes[0,1].plot(T_vals, F, 'g'); axes[0,1].set_title("Helmholtz Free Energy F(T)")
axes[0,1].set_xlabel("T"); axes[0,1].set_ylabel("F"); axes[0,1].grid(alpha=.3)

axes[1,0].plot(T_vals, S, 'r'); axes[1,0].set_title("Entropy S(T)")
axes[1,0].set_xlabel("T"); axes[1,0].set_ylabel("S"); axes[1,0].grid(alpha=.3)

axes[1,1].plot(T_vals, Cv, 'm', label=r"$C_V$")
axes[1,1].axhline(kB, color='k', linestyle='--', label=r"Classical $k_B$")
axes[1,1].set_title("Heat Capacity $C_V(T)$")
axes[1,1].set_xlabel("T"); axes[1,1].set_ylabel(r"$C_V$")
axes[1,1].legend(); axes[1,1].grid(alpha=.3)

plt.tight_layout()
plt.show()

# Verify U = F + TS
err = np.max(np.abs(U - (F + T_vals * S)))
print(f"Max |U - (F+TS)| = {err:.2e}  (should be ~0)")
```

# Q. Maximum Entropy Distribution under Mean and Variance Constraints

We seek the pdf $p(x)$ on $(-\infty,\infty)$ that **maximises** the differential entropy

$$H[p]=-\int_{-\infty}^\infty p(x)\ln p(x)\,dx,$$

subject to the three constraints:

$$\int p(x)\,dx=1,\quad \int x\,p(x)\,dx=\mu,\quad \int (x-\mu)^2 p(x)\,dx=\sigma^2.$$

---
## Solution via the Calculus of Variations (Lagrange multipliers)

Introduce Lagrange multipliers $\lambda_0,\lambda_1,\lambda_2$ and form the functional

$$\mathcal{L}[p]=\int\!\left[-p\ln p - \lambda_0 p - \lambda_1 x p - \lambda_2(x-\mu)^2 p\right]dx.$$

Setting the functional derivative to zero, $\delta\mathcal{L}/\delta p=0$:

$$-\ln p(x)-1-\lambda_0-\lambda_1 x -\lambda_2(x-\mu)^2=0,$$

which gives

$$\ln p(x)=-1-\lambda_0-\lambda_1 x -\lambda_2(x-\mu)^2,$$

i.e.

$$p(x)=\exp\!\left(-1-\lambda_0-\lambda_1 x-\lambda_2(x-\mu)^2\right).$$

---
## Fixing the multipliers from the constraints

**Step 1 — normalisability requires $\lambda_2>0$.**  Only then does $\int e^{-\lambda_2(x-\mu)^2}\,dx$ converge.

**Step 2 — mean constraint $\mathbb E[X]=\mu$.**  The distribution has the form $p(x)\propto e^{-\lambda_1 x-\lambda_2(x-\mu)^2}$.  Complete the square in $x$:

$$\lambda_1 x+\lambda_2(x-\mu)^2=\lambda_2\!\left(x-\mu+\frac{\lambda_1}{2\lambda_2}\right)^2-\frac{\lambda_1^2}{4\lambda_2}+\lambda_1\mu.$$

For the mean to equal $\mu$ we need the centre of the Gaussian to be at $\mu$, which forces

$$\frac{\lambda_1}{2\lambda_2}=0 \implies \lambda_1=0.$$

**Step 3 — variance constraint $\mathbb E[(X-\mu)^2]=\sigma^2$.**  With $\lambda_1=0$ we have $p(x)\propto e^{-\lambda_2(x-\mu)^2}$.  The normalised Gaussian with this form is $p(x)=\sqrt{\lambda_2/\pi}\,e^{-\lambda_2(x-\mu)^2}$.  Its variance is $1/(2\lambda_2)$, so

$$\frac{1}{2\lambda_2}=\sigma^2\implies\lambda_2=\frac{1}{2\sigma^2}.$$

**Step 4 — normalisation constant.**  With $\lambda_2=1/(2\sigma^2)$ and $\lambda_1=0$:

$$e^{-1-\lambda_0}=\frac{1}{\sqrt{2\pi\sigma^2}}.$$

---
## Result

Substituting back:

$$\boxed{p(x)=\frac{1}{\sqrt{2\pi\sigma^2}}\exp\!\left(-\frac{(x-\mu)^2}{2\sigma^2}\right).}$$

The maximum-entropy distribution on $\mathbb{R}$ with fixed mean $\mu$ and variance $\sigma^2$ is the **Gaussian (Normal) distribution** $\mathcal{N}(\mu,\sigma^2)$. $\blacksquare$

The maximum entropy value is $H^* = \tfrac12\ln(2\pi e\sigma^2)$.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, laplace, t as tdist

# Compare entropy of Gaussian vs other distributions with same mean & variance
mu, sigma2 = 0.0, 1.0
sigma = np.sqrt(sigma2)
x = np.linspace(-5, 5, 2000)
dx = x[1] - x[0]

# Gaussian (MaxEnt)
p_gauss = norm.pdf(x, mu, sigma)

# Laplace with same mean and variance: scale b = sigma/sqrt(2)
b_lap = sigma / np.sqrt(2)
p_lap = laplace.pdf(x, mu, b_lap)

# Differential entropy H = -int p ln p dx
def entropy(p, dx):
    mask = p > 0
    return -np.sum(p[mask] * np.log(p[mask])) * dx

H_gauss = entropy(p_gauss, dx)
H_lap   = entropy(p_lap,   dx)
print(f"Gaussian entropy  H = {H_gauss:.4f}  (theory: {0.5*np.log(2*np.pi*np.e*sigma2):.4f})")
print(f"Laplace entropy   H = {H_lap:.4f}  (< Gaussian, confirming MaxEnt result)")

plt.figure(figsize=(7, 4))
plt.plot(x, p_gauss, 'b-', linewidth=2, label=f"Gaussian (MaxEnt) H={H_gauss:.3f}")
plt.plot(x, p_lap,   'r--', linewidth=2, label=f"Laplace (same var) H={H_lap:.3f}")
plt.title(r"MaxEnt: Gaussian maximises $H$ for fixed $\mu,\sigma^2$")
plt.xlabel("x"); plt.ylabel("p(x)"); plt.legend(); plt.grid(alpha=.3)
plt.show()
```

# Q. Multiclass Logistic Regression: Automatic Classification of Emails

## Setup

Classes: $Y\in\{0,1\}$ (Finance, Technical). Feature vector $\mathbf x=(x_1,x_2,x_3)^T\in\{0,1\}^3$.

**Observed scenarios and their marginal probabilities:**

| Scenario | $\mathbf x$ | $P(\mathbf x)$ |
|---|---|---|
| A (Pure Financial) | $(1,0,0)$ | 0.40 |
| B (Pure Access)    | $(0,1,0)$ | 0.25 |
| C (Pure Technical) | $(0,0,1)$ | 0.05 |
| D (Access+Tech)    | $(0,1,1)$ | 0.15 |
| E (Billing+Access) | $(1,1,0)$ | 0.05 |

Unobserved mass: $1-(0.40+0.25+0.05+0.15+0.05)=0.10$ (other scenarios not listed).

**Constraints (observed joint expectations from audits):**

$$\mathbb E[x_1\cdot\mathbb{1}(Y=1)]=0.05,\quad
\mathbb E[x_2\cdot\mathbb{1}(Y=1)]=0.15,\quad
\mathbb E[x_3\cdot\mathbb{1}(Y=1)]=0.16.$$

---
## Maximum Entropy Formulation

We want to find the **conditional distribution** $P(Y=1|\mathbf x)$ that satisfies the constraints and is otherwise maximally unbiased (maximum entropy).

The MaxEnt solution for a binary conditional distribution given linear feature constraints takes the **logistic regression** form:

$$P(Y=1\mid\mathbf x)=\sigma(\theta_0+\theta_1 x_1+\theta_2 x_2+\theta_3 x_3)
=\frac{1}{1+e^{-(\theta_0+\boldsymbol\theta^T\mathbf x)}},$$

where $\sigma(z)=1/(1+e^{-z})$ is the sigmoid. This is precisely the MaxEnt distribution over $\{0,1\}$ with linear sufficient statistics $\{1,x_1,x_2,x_3\}$.

---
## Solving for the Parameters

The constraints translate to a nonlinear system. Using the five listed scenarios (which account for 0.90 of the total probability mass; assuming $P(Y=1|\mathbf x_{\text{other}})$ is consistent with the constraints) we solve numerically.

The constraint equations are:

$$\sum_i P(\mathbf x_i)\,x_{1,i}\,\sigma(\boldsymbol\theta^T\tilde{\mathbf x}_i)=0.05,$$
$$\sum_i P(\mathbf x_i)\,x_{2,i}\,\sigma(\boldsymbol\theta^T\tilde{\mathbf x}_i)=0.15,$$
$$\sum_i P(\mathbf x_i)\,x_{3,i}\,\sigma(\boldsymbol\theta^T\tilde{\mathbf x}_i)=0.16,$$

where $\tilde{\mathbf x}_i=(1,x_{1,i},x_{2,i},x_{3,i})^T$ includes the bias. We solve this 3-constraint / 4-parameter system by also imposing a **zero-mean prior** (regularisation: $\theta_0$ set by overall base rate) or by fixing one degree of freedom.


```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

# -------- Data: scenarios, marginal P(x), feature vectors --------
scenarios = {
    'A (Financial)':     {'x': np.array([1,0,0]), 'Px': 0.40},
    'B (Access)':        {'x': np.array([0,1,0]), 'Px': 0.25},
    'C (Technical)':     {'x': np.array([0,0,1]), 'Px': 0.05},
    'D (Access+Tech)':   {'x': np.array([0,1,1]), 'Px': 0.15},
    'E (Billing+Access)':{'x': np.array([1,1,0]), 'Px': 0.05},
}
# Constraints
targets = np.array([0.05, 0.15, 0.16])

def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-np.clip(z, -500, 500)))

def compute_expectations(theta):
    # E[x_j * 1(Y=1)] for j=1,2,3 under the logistic model.
    E = np.zeros(3)
    for s in scenarios.values():
        x = s['x']
        Px = s['Px']
        # augmented: bias + features
        z = theta[0] + theta[1]*x[0] + theta[2]*x[1] + theta[3]*x[2]
        p_tech = sigmoid(z)
        E += Px * x * p_tech
    return E

def equations(theta):
    return compute_expectations(theta) - targets

# Solve: we have 3 equations, 4 unknowns; fix theta[0] (bias) by overall marginal P(Y=1)
# P(Y=1) can be estimated: sum of constraints (they partition by feature, not exhaustive)
# Use a grid search over theta[0], then solve the 3-equation system for theta[1:4]

best = None
best_err = np.inf
for t0 in np.linspace(-3, 3, 30):
    def eq3(theta3):
        th = np.array([t0, theta3[0], theta3[1], theta3[2]])
        return equations(th)
    try:
        sol = fsolve(eq3, [0.0, 0.5, 1.5], full_output=True)
        if sol[2] == 1:
            err = np.max(np.abs(eq3(sol[0])))
            if err < best_err:
                best_err = err
                best = np.array([t0, sol[0][0], sol[0][1], sol[0][2]])
    except:
        pass

theta_star = best
print(f"Optimal theta: bias={theta_star[0]:.4f}, theta1={theta_star[1]:.4f}, "
      f"theta2={theta_star[2]:.4f}, theta3={theta_star[3]:.4f}")
print(f"Constraint residuals: {equations(theta_star)}")

# ---- Print P(Y=1|x) for each scenario ----
print("\nP(Y=1 | x) for each scenario:")
for name, s in scenarios.items():
    x = s['x']
    z = theta_star[0] + theta_star[1]*x[0] + theta_star[2]*x[1] + theta_star[3]*x[2]
    p1 = sigmoid(z)
    print(f"  {name:25s}: P(Tech)={p1:.4f}, P(Finance)={1-p1:.4f}")

# ---- Visualise ----
names = list(scenarios.keys())
p_tech = []
for s in scenarios.values():
    x = s['x']
    z = theta_star[0] + theta_star[1]*x[0] + theta_star[2]*x[1] + theta_star[3]*x[2]
    p_tech.append(sigmoid(z))

fig, ax = plt.subplots(figsize=(8, 4))
bars = ax.bar(names, p_tech, color=['#e74c3c','#3498db','#2ecc71','#9b59b6','#f39c12'])
ax.set_ylabel("P(Y=1 | Technical)")
ax.set_title("MaxEnt Logistic Regression: Routing Probabilities")
ax.set_ylim(0, 1)
for bar, p in zip(bars, p_tech):
    ax.text(bar.get_x()+bar.get_width()/2, p+0.02, f"{p:.3f}", ha='center', fontsize=10)
plt.xticks(rotation=20, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

# Q. Asset Pricing and Portfolio Optimization via MaxEnt

## Problem Setup

Find the **maximum entropy** pdf $p(S_T)$ for $S_T>0$ subject to:

| Constraint | Value |
|---|---|
| $\int_0^\infty p(s)\,ds = 1$ | normalisation |
| $\mathbb E[S_T] = 102$ | risk-free forward price |
| $\mathbb E[S_T^2] = 10{,}600$ | market volatility |
| $\mathbb E[\ln S_T] = 4.60$ | downside skew (put option pricing) |

---
## Derivation

Using Lagrange multipliers $\lambda_0,\lambda_1,\lambda_2,\lambda_3$ for the four constraints, the MaxEnt functional is

$$\mathcal L[p] = -\int_0^\infty p\ln p\,ds
- \lambda_0\!\int p\,ds
- \lambda_1\!\int sp\,ds
- \lambda_2\!\int s^2 p\,ds
- \lambda_3\!\int (\ln s)p\,ds.$$

Setting $\delta\mathcal L/\delta p=0$:

$$-\ln p(s)-1-\lambda_0-\lambda_1 s-\lambda_2 s^2-\lambda_3\ln s=0,$$

$$\boxed{p(s)=\frac{1}{Z}\,s^{-\lambda_3}\exp\!\left(-\lambda_1 s-\lambda_2 s^2\right),\quad s>0,}$$

where $Z=e^{1+\lambda_0}$ is the normalisation constant.

This is a **generalised Gamma / skewed distribution**: the $s^{-\lambda_3}$ power law (from the log constraint) introduces the downside skew (heavier left tail relative to a log-normal), while $e^{-\lambda_2 s^2}$ provides the Gaussian tail cut-off.

---
## Interpretation of Lagrange Multipliers

| Multiplier | Constraint it enforces | Economic meaning |
|---|---|---|
| $\lambda_0$ | Normalisation | Calibrates overall scale |
| $\lambda_1$ | $\mathbb E[S_T]=102$ | Risk-free drift / forward price |
| $\lambda_2$ | $\mathbb E[S_T^2]=10{,}600$ | Market volatility ($\text{Var}=10{,}600-102^2=196,\ \sigma=14$) |
| $\lambda_3$ | $\mathbb E[\ln S_T]=4.60$ | Skew / put option implied crash risk |

Note: $\text{Var}(S_T)=\mathbb E[S_T^2]-(\mathbb E[S_T])^2=10600-10404=196$, so $\sigma_{S_T}=14$ (≈14% of the $100 price).


```python
import numpy as np
from scipy.optimize import fsolve
from scipy.integrate import quad
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# ---- Target moments ----
E_S   = 102.0
E_S2  = 10600.0
E_lnS = 4.60

def p_unnorm(s, lam1, lam2, lam3):
    # Unnormalised MaxEnt density: s^{-lam3} * exp(-lam1*s - lam2*s^2)
    if s <= 0:
        return 0.0
    log_p = -lam3 * np.log(s) - lam1 * s - lam2 * s**2
    return np.exp(np.clip(log_p, -700, 700))

def moments(lam1, lam2, lam3, s_max=500.0):
    Z,   _ = quad(lambda s: p_unnorm(s, lam1, lam2, lam3),    1e-6, s_max)
    M1,  _ = quad(lambda s: s*p_unnorm(s, lam1, lam2, lam3),  1e-6, s_max)
    M2,  _ = quad(lambda s: s**2*p_unnorm(s, lam1, lam2, lam3),1e-6, s_max)
    Mln, _ = quad(lambda s: np.log(s)*p_unnorm(s, lam1, lam2, lam3),1e-6, s_max)
    return Z, M1/Z, M2/Z, Mln/Z

def equations(params):
    lam1, lam2, lam3 = params
    if lam2 <= 0:
        return [1e6, 1e6, 1e6]
    try:
        Z, m1, m2, mln = moments(lam1, lam2, lam3)
        return [m1 - E_S, m2 - E_S2, mln - E_lnS]
    except:
        return [1e6, 1e6, 1e6]

# Solve the system — good initial guess from log-normal approximation
# For log-normal: mu_ln ~ ln(E_S) - sigma_ln^2/2, sigma_ln^2 ~ ln(E_S2/E_S^2)
sigma_ln2 = np.log(E_S2 / E_S**2)
mu_ln = np.log(E_S) - sigma_ln2/2
print(f"Log-normal initial approximation: mu_ln={mu_ln:.4f}, sigma_ln={np.sqrt(sigma_ln2):.4f}")
print(f"Check E[ln S_T] from log-normal: {mu_ln:.4f} (target {E_lnS})")

sol = fsolve(equations, x0=[0.001, 0.00004, -1.0], full_output=True)
lam1_star, lam2_star, lam3_star = sol[0]
residuals = equations(sol[0])

print(f"\nOptimal Lagrange multipliers:")
print(f"  lambda1 = {lam1_star:.6f}  (E[S] constraint)")
print(f"  lambda2 = {lam2_star:.6f}  (E[S^2] constraint)")
print(f"  lambda3 = {lam3_star:.6f}  (E[ln S] constraint)")
print(f"Residuals: E[S]-102={residuals[0]:.4f}, E[S^2]-10600={residuals[1]:.2f}, E[lnS]-4.60={residuals[2]:.4f}")

# ---- Compute and plot the density ----
Z_star, m1, m2, mln = moments(lam1_star, lam2_star, lam3_star)

s_vals = np.linspace(1, 250, 1000)
p_vals = np.array([p_unnorm(s, lam1_star, lam2_star, lam3_star)/Z_star for s in s_vals])

# Log-normal with same mean and variance (for comparison)
from scipy.stats import lognorm
sigma_ln = np.sqrt(np.log(E_S2/E_S**2))
mu_ln2   = np.log(E_S) - sigma_ln**2/2
p_lognorm = lognorm.pdf(s_vals, s=sigma_ln, scale=np.exp(mu_ln2))

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(s_vals, p_vals, 'b-', lw=2, label='MaxEnt $p(S_T)$')
axes[0].plot(s_vals, p_lognorm, 'r--', lw=2, label='Log-Normal (same E[S],Var[S])')
axes[0].axvline(E_S, color='k', linestyle=':', label=f'$E[S_T]={E_S}$')
axes[0].set_xlabel('$S_T$'); axes[0].set_ylabel('$p(S_T)$')
axes[0].set_title('MaxEnt vs Log-Normal Risk-Neutral Distribution')
axes[0].legend(); axes[0].grid(alpha=.3)
axes[0].set_xlim(0, 250)

# Log-scale plot to show fat left tail
axes[1].semilogy(s_vals, p_vals, 'b-', lw=2, label='MaxEnt $p(S_T)$')
axes[1].semilogy(s_vals, p_lognorm, 'r--', lw=2, label='Log-Normal')
axes[1].set_xlabel('$S_T$'); axes[1].set_ylabel('$p(S_T)$ (log scale)')
axes[1].set_title('Left-tail comparison (log scale)')
axes[1].legend(); axes[1].grid(alpha=.3)
axes[1].set_xlim(0, 250)

plt.tight_layout()
plt.show()

print(f"\nVerified moments under MaxEnt distribution:")
print(f"  E[S_T]      = {m1:.3f}  (target: {E_S})")
print(f"  E[S_T^2]    = {m2:.1f}  (target: {E_S2})")
print(f"  E[ln(S_T)]  = {mln:.4f}  (target: {E_lnS})")
print(f"  Var(S_T)    = {m2-m1**2:.3f}  => std = {np.sqrt(m2-m1**2):.3f}")
```
