## Q. Analytical Derivation

**Model**

$$x_k^- = A_{k-1}x_{k-1}^+ + G_{k-1}w_{k-1}, \qquad y_k^- = H_k x_k^- + z_k,$$

where $x_{k-1}^+ \sim \mathcal N(m_{k-1},P_{k-1})$, $w_{k-1}\sim\mathcal N(0,\Sigma_p)$, $z_k\sim\mathcal N(0,\Sigma_m)$, all mutually independent.

### 1. Prior (predictive) distribution of $x_k^-$

Since $x_k^-$ is an affine combination of two **independent** Gaussian vectors $x_{k-1}^+$ and $w_{k-1}$, it is itself Gaussian. Affine transformations of Gaussians remain Gaussian, and the mean/covariance transform as:

$$\mathbb E[x_k^-] = A_{k-1}\,\mathbb E[x_{k-1}^+] + G_{k-1}\,\mathbb E[w_{k-1}] = A_{k-1}m_{k-1},$$

since $\mathbb E[w_{k-1}]=0$. For the covariance, using independence of $x_{k-1}^+$ and $w_{k-1}$ (so the cross-covariance term vanishes):

$$\operatorname{Cov}(x_k^-) = A_{k-1}\operatorname{Cov}(x_{k-1}^+)A_{k-1}^T + G_{k-1}\operatorname{Cov}(w_{k-1})G_{k-1}^T.$$

Hence

$$x_k^- \sim \mathcal N(m_k^-, P_k^-), \qquad
m_k^- = A_{k-1}m_{k-1}, \qquad
P_k^- = A_{k-1}P_{k-1}A_{k-1}^T + G_{k-1}\Sigma_p G_{k-1}^T. \qquad \blacksquare$$

### 2. Distribution of $y_k^-$

$y_k^- = H_k x_k^- + z_k$ is an affine map of the independent Gaussians $x_k^-\sim\mathcal N(m_k^-,P_k^-)$ and $z_k\sim\mathcal N(0,\Sigma_m)$, so

$$\mathbb E[y_k^-] = H_k m_k^-, \qquad
\operatorname{Cov}(y_k^-) = H_kP_k^-H_k^T + \Sigma_m,$$

i.e.

$$y_k^- \sim \mathcal N\!\left(H_k m_k^-,\; H_kP_k^-H_k^T+\Sigma_m\right). \qquad \blacksquare$$

### 3. Joint distribution of $(x_k^-,y_k^-)$

The pair $\big(x_k^-,\,y_k^-\big)$ is an affine function of the jointly-Gaussian, independent vector $(x_k^-,z_k)$, hence is jointly Gaussian. The means are $m_k^-$ and $H_km_k^-$ as above. The cross-covariance is

$$\operatorname{Cov}(x_k^-,y_k^-) = \operatorname{Cov}\big(x_k^-,\,H_kx_k^-+z_k\big)
= \operatorname{Cov}(x_k^-)H_k^T = P_k^-H_k^T,$$

(using $\operatorname{Cov}(x_k^-,z_k)=0$). Therefore

$$\begin{bmatrix}x_k^-\\ y_k^-\end{bmatrix}
\sim \mathcal N\!\left(
\begin{bmatrix}m_k^-\\ H_km_k^-\end{bmatrix},\;
\begin{bmatrix}P_k^- & P_k^-H_k^T\\ H_kP_k^- & H_kP_k^-H_k^T+\Sigma_m\end{bmatrix}
\right). \qquad \blacksquare$$

### 4. Conditional (posterior) distribution

For a jointly Gaussian vector $\begin{bmatrix}x\\y\end{bmatrix}\sim\mathcal N\!\left(\begin{bmatrix}\mu_x\\\mu_y\end{bmatrix},\begin{bmatrix}\Sigma_{xx}&\Sigma_{xy}\\\Sigma_{yx}&\Sigma_{yy}\end{bmatrix}\right)$, the standard Gaussian conditioning identity gives

$$x\mid y=y_{\text{obs}} \sim \mathcal N\!\left(\mu_x+\Sigma_{xy}\Sigma_{yy}^{-1}(y_{\text{obs}}-\mu_y),\;\; \Sigma_{xx}-\Sigma_{xy}\Sigma_{yy}^{-1}\Sigma_{yx}\right).$$

Substituting $\mu_x=m_k^-$, $\mu_y=H_km_k^-$, $\Sigma_{xx}=P_k^-$, $\Sigma_{xy}=P_k^-H_k^T$, $\Sigma_{yy}=H_kP_k^-H_k^T+\Sigma_m$, and defining

$$K_k \triangleq P_k^-H_k^T\big(H_kP_k^-H_k^T+\Sigma_m\big)^{-1},$$

we obtain

$$x_k^+ \triangleq (x_k^- \mid y_k^- = y_k^{\text{obs}}) \sim \mathcal N(m_k,P_k),$$

$$m_k = m_k^- + K_k\big(y_k^{\text{obs}}-H_km_k^-\big), \qquad
P_k = P_k^- - K_kH_kP_k^- = (I-K_kH_k)P_k^-. \qquad \blacksquare$$

### 5. Conditional mean and variance

Directly from Part 4:

$$\mathbb E\big[x_k^- \mid y_k^-=y_k^{\text{obs}}\big] = m_k = m_k^- + K_k(y_k^{\text{obs}}-H_km_k^-),$$

$$\operatorname{Var}\big(x_k^- \mid y_k^-=y_k^{\text{obs}}\big) = P_k = (I-K_kH_k)P_k^-.$$

## Q. 1-D Example

**Model:**
$$x_k^- = a\,x_{k-1}^+ + w_{k-1},\quad w_{k-1}\sim\mathcal N(0,q), \qquad
y_k^- = h\,x_k^- + z_k,\quad z_k\sim\mathcal N(0,r),$$
with $x_0\sim\mathcal N(m_0,P_0)$. These are scalar special cases ($A_{k-1}=a$, $G_{k-1}=1$, $\Sigma_p=q$, $H_k=h$, $\Sigma_m=r$) of the general results derived above.

### 1. Prior update
Substituting $A_{k-1}=a,\, G_{k-1}=1,\,\Sigma_p=q$ into Part 1 of the general derivation:

$$m_k^- = a\,m_{k-1}, \qquad P_k^- = a^2P_{k-1}+q. \qquad \blacksquare$$

### 2. Posterior update
With $H_k=h,\ \Sigma_m=r$, the Kalman gain from Part 4 becomes the scalar

$$K_k = \frac{P_k^-h}{h^2P_k^-+r} = \frac{P_k^-h}{S_k}, \qquad S_k \triangleq h^2P_k^-+r.$$

Let $v_k \triangleq y_k^{\text{obs}}-hm_k^-$ (the *innovation*). Then

$$m_k = m_k^- + K_kv_k = m_k^- + \frac{P_k^-h}{S_k}\big(y_k^{\text{obs}}-hm_k^-\big),$$

$$P_k = (1-K_kh)P_k^- = \left(1-\frac{P_k^-h^2}{S_k}\right)P_k^-. \qquad \blacksquare$$

### 3. Predictive measurement distribution
Special case of the general Part 2 with $H_k=h,\Sigma_m=r$:

$$p(y_k^-\mid Y_{k-1}) = \mathcal N\!\left(hm_k^-,\; h^2P_k^-+r\right). \qquad \blacksquare$$

### 4. Posterior-predictive measurement distribution
After conditioning on $y_k^{\text{obs}}$, the (filtered) state is $x_k^+\sim\mathcal N(m_k,P_k)$. Pushing this through the (noisy) measurement map $y=hx+z$ exactly as in Part 2 (but using the posterior $m_k,P_k$ instead of the prior $m_k^-,P_k^-$) gives

$$p(y_k^- \mid Y_k) = \mathcal N\!\left(hm_k,\; h^2P_k+r\right). \qquad \blacksquare$$

### 5. Numerical animation of the prior and posterior distributions
We pick $a=0.9$ (mild mean-reversion), $q=0.1$ (process noise), $h=1$ (direct observation), $r=0.5$ (fairly noisy sensor), simulate a true trajectory and noisy measurements, run the scalar Kalman filter, and animate $p(y_k^-\mid Y_{k-1})$ (prior-predictive) vs. $p(y_k^-\mid Y_k)$ (posterior-predictive) at each step.


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from scipy.stats import norm
from IPython.display import HTML

# ---- 1-D Kalman filter parameters ----
a, q, h, r = 0.9, 0.1, 1.0, 0.5
m0, P0 = 0.0, 1.0
N = 30
rng = np.random.default_rng(0)

# Simulate true state and noisy measurements
x_true = np.zeros(N)
x_true[0] = m0 + np.sqrt(P0) * rng.standard_normal()
for k in range(1, N):
    x_true[k] = a * x_true[k-1] + np.sqrt(q) * rng.standard_normal()
y_obs = h * x_true + np.sqrt(r) * rng.standard_normal(N)

# Run the scalar Kalman filter, storing prior/posterior stats
m_prior, P_prior = np.zeros(N), np.zeros(N)
m_post,  P_post  = np.zeros(N), np.zeros(N)

m_prev, P_prev = m0, P0
for k in range(N):
    # --- predict ---
    mk_minus = a * m_prev
    Pk_minus = a**2 * P_prev + q
    m_prior[k], P_prior[k] = mk_minus, Pk_minus

    # --- update ---
    Sk = h**2 * Pk_minus + r
    Kk = Pk_minus * h / Sk
    mk = mk_minus + Kk * (y_obs[k] - h * mk_minus)
    Pk = (1 - Kk * h) * Pk_minus
    m_post[k], P_post[k] = mk, Pk

    m_prev, P_prev = mk, Pk

print("Filter run complete. Final estimate m_N =", m_post[-1], " P_N =", P_post[-1])
```

```python
# ---- Animate prior-predictive vs posterior-predictive measurement distributions ----
fig, ax = plt.subplots(figsize=(7, 4.5))
y_grid = np.linspace(-6, 6, 600)

def update(k):
    ax.clear()
    mu_prior = h * m_prior[k]
    var_prior = h**2 * P_prior[k] + r
    mu_post = h * m_post[k]
    var_post = h**2 * P_post[k] + r

    ax.plot(y_grid, norm.pdf(y_grid, mu_prior, np.sqrt(var_prior)),
            label=r'$p(y_k^-\mid Y_{k-1})$ (prior-predictive)', color='tab:blue')
    ax.plot(y_grid, norm.pdf(y_grid, mu_post, np.sqrt(var_post)),
            label=r'$p(y_k^-\mid Y_k)$ (posterior-predictive)', color='tab:green')
    ax.axvline(y_obs[k], color='tab:red', linestyle='--', label=r'$y_k^{obs}$')
    ax.set_title(f"Step k = {k}")
    ax.set_xlabel("y")
    ax.set_ylabel("density")
    ax.set_ylim(0, 1.4)
    ax.legend(loc='upper right')
    return []

anim = FuncAnimation(fig, update, frames=N, interval=600, blit=False)
plt.close(fig)
HTML(anim.to_jshtml())
```

## Q. 2D-Position Estimation

### Part A — Derivation of $A$, $H$, $G$

The hidden state is $x_k=\big[p_x(k),\,p_y(k),\,v_x(k),\,v_y(k)\big]^T$. We model constant-velocity motion over a sampling interval $\Delta t$, where the *true* (unmodeled) acceleration is treated as random process noise $w=(w_x,w_y)\sim\mathcal N(0,\Sigma_p)$ acting over the interval. Using the standard discretization of a constant-velocity / white-noise-acceleration model, integrating $\dot p = v,\ \dot v = w$ over $[t_{k-1},t_k]$ with $\Delta t = t_k-t_{k-1}$ gives, component-wise,

$$p_x(k) = p_x(k-1) + \Delta t\, v_x(k-1) + \tfrac12\Delta t^2\, w_x,$$
$$p_y(k) = p_y(k-1) + \Delta t\, v_y(k-1) + \tfrac12\Delta t^2\, w_y,$$
$$v_x(k) = v_x(k-1) + \Delta t\, w_x, \qquad v_y(k) = v_y(k-1) + \Delta t\, w_y.$$

Writing this in matrix form $x_k^- = A x_{k-1}^+ + G w_{k-1}$ with $x_{k-1}^+=[p_x,p_y,v_x,v_y]^T_{k-1}$ and $w_{k-1}=[w_x,w_y]^T$:

$$
A=\begin{bmatrix}1&0&\Delta t&0\\0&1&0&\Delta t\\0&0&1&0\\0&0&0&1\end{bmatrix},
\qquad
G=\begin{bmatrix}\tfrac12\Delta t^2&0\\0&\tfrac12\Delta t^2\\ \Delta t&0\\0&\Delta t\end{bmatrix}.
$$

Indeed, multiplying out $Ax_{k-1}^++Gw_{k-1}$ reproduces exactly the four equations above, row by row. $\blacksquare$

For the measurement, only position is observed: $y_k=\big[p_x^{\text{meas}}(k),\,p_y^{\text{meas}}(k)\big]^T = Hx_k^+ + z_k$. Since the measurement simply *selects* the first two state components,

$$
H=\begin{bmatrix}1&0&0&0\\0&1&0&0\end{bmatrix},
$$

because $Hx_k^+=[p_x(k),p_y(k)]^T$ exactly matches the measured quantities. $\blacksquare$


### Part B — Python implementation: filtering noisy GPS position measurements

Below is a complete, self-contained Kalman filter implementation for the 2D constant-velocity model derived in Part A. It:

1. Simulates a true 2D trajectory and corrupts the position with GPS-like measurement noise.
2. Implements the **predict** and **update** steps using the matrices $A$, $H$, $G$ derived above.
3. Plots the true path, noisy GPS measurements, and the filtered (estimated) path.


```python
import numpy as np
import matplotlib.pyplot as plt

class KalmanFilter2D:
    '''Constant-velocity 2D Kalman filter for GPS position tracking.'''

    def __init__(self, dt, sigma_p, sigma_m, m0=None, P0=None):
        self.dt = dt
        # State transition matrix A
        self.A = np.array([
            [1, 0, dt, 0],
            [0, 1, 0, dt],
            [0, 0, 1, 0],
            [0, 0, 0, 1]
        ])
        # Measurement matrix H (observe position only)
        self.H = np.array([
            [1, 0, 0, 0],
            [0, 1, 0, 0]
        ])
        # Process noise input matrix G
        self.G = np.array([
            [0.5 * dt**2, 0],
            [0, 0.5 * dt**2],
            [dt, 0],
            [0, dt]
        ])
        # Process noise covariance (acceleration noise), Sigma_p (2x2)
        self.Sigma_p = sigma_p * np.eye(2)
        # Measurement noise covariance, Sigma_m (2x2), e.g. GPS noise
        self.Sigma_m = sigma_m * np.eye(2)

        # Initial state estimate
        self.m = np.zeros(4) if m0 is None else np.array(m0, dtype=float)
        self.P = np.eye(4) * 10.0 if P0 is None else np.array(P0, dtype=float)

    def predict(self):
        '''Time update (prior).'''
        self.m = self.A @ self.m
        self.P = self.A @ self.P @ self.A.T + self.G @ self.Sigma_p @ self.G.T
        return self.m.copy(), self.P.copy()

    def update(self, y_obs):
        '''Measurement update (posterior) given an observed GPS fix y_obs = [px, py].'''
        y_obs = np.asarray(y_obs, dtype=float)
        S = self.H @ self.P @ self.H.T + self.Sigma_m          # innovation covariance
        K = self.P @ self.H.T @ np.linalg.inv(S)                # Kalman gain
        innovation = y_obs - self.H @ self.m
        self.m = self.m + K @ innovation
        self.P = (np.eye(4) - K @ self.H) @ self.P
        return self.m.copy(), self.P.copy()

    def step(self, y_obs=None):
        '''Convenience: run predict, then update if a measurement is supplied.'''
        self.predict()
        if y_obs is not None:
            self.update(y_obs)
        return self.m.copy(), self.P.copy()
```

```python
# ---- Simulate a true trajectory and noisy GPS measurements ----
np.random.seed(42)

dt = 1.0          # seconds between GPS fixes
N = 60             # number of time steps
true_sigma_p = 0.05   # true process (acceleration) noise variance
gps_sigma_m = 9.0     # true GPS measurement noise variance (~3 m std dev)

# Build the same A, G used by the filter to *generate* ground truth + noisy measurements
A_true = np.array([[1,0,dt,0],[0,1,0,dt],[0,0,1,0],[0,0,0,1]])
G_true = np.array([[0.5*dt**2,0],[0,0.5*dt**2],[dt,0],[0,dt]])
H_true = np.array([[1,0,0,0],[0,1,0,0]])

x_true = np.zeros((N, 4))
x_true[0] = [0, 0, 1.0, 0.5]   # start at origin, moving with constant velocity
for k in range(1, N):
    w = np.random.multivariate_normal([0,0], true_sigma_p*np.eye(2))
    x_true[k] = A_true @ x_true[k-1] + G_true @ w

y_gps = np.zeros((N, 2))
for k in range(N):
    z = np.random.multivariate_normal([0,0], gps_sigma_m*np.eye(2))
    y_gps[k] = H_true @ x_true[k] + z
```

```python
# ---- Run the Kalman filter on the noisy GPS measurements ----
kf = KalmanFilter2D(dt=dt, sigma_p=true_sigma_p, sigma_m=gps_sigma_m,
                     m0=[y_gps[0,0], y_gps[0,1], 0, 0], P0=np.eye(4)*5)

estimates = np.zeros((N, 4))
covariances = []
for k in range(N):
    m, P = kf.step(y_gps[k])
    estimates[k] = m
    covariances.append(P)

# ---- Plot results ----
fig, ax = plt.subplots(figsize=(7, 7))
ax.plot(x_true[:,0], x_true[:,1], 'g-', linewidth=2, label='True trajectory')
ax.scatter(y_gps[:,0], y_gps[:,1], c='red', s=20, alpha=0.6, label='Noisy GPS measurements')
ax.plot(estimates[:,0], estimates[:,1], 'b-', linewidth=2, label='Kalman filter estimate')
ax.set_xlabel('p_x (m)')
ax.set_ylabel('p_y (m)')
ax.set_title('2D Constant-Velocity Kalman Filter: GPS Position Tracking')
ax.legend()
ax.axis('equal')
ax.grid(alpha=0.3)
plt.show()

rmse_gps = np.sqrt(np.mean(np.sum((y_gps - x_true[:,:2])**2, axis=1)))
rmse_kf  = np.sqrt(np.mean(np.sum((estimates[:,:2] - x_true[:,:2])**2, axis=1)))
print(f"RMSE of raw GPS measurements vs truth: {rmse_gps:.3f} m")
print(f"RMSE of Kalman filter estimate vs truth: {rmse_kf:.3f} m")
```

As expected, the Kalman filter estimate (blue) tracks the true trajectory (green) much more closely than the raw noisy GPS fixes (red), and the RMSE printed above is substantially lower for the filtered estimate — demonstrating the variance-reduction benefit of fusing the constant-velocity motion model with noisy position measurements via the equations derived in Parts A and the Analytical Derivation section.
