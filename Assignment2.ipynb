# Q3) Transformer Failures

### 1. Probability Space $(\Omega, \mathfrak{F}, P)$
- **Sample Space:** $\Omega = \{0, 1, 2, 3\}$
- **$\sigma$-algebra:** $\mathfrak{F} = \mathfrak{P}(\Omega)$ (Power set of $\Omega$, containing $2^4 = 16$ events).
- **Probability Measure:** $P(\{0\}) = 0.50, P(\{1\}) = 0.30, P(\{2\}) = 0.15, P(\{3\}) = 0.05$.

### 2. Random Variable $X$
$X$ is defined as the identity map $X(\omega) = \omega$ for $\omega \in \Omega$.

### 3. Probability Mass Function (PMF)
$$p_X(x) = \begin{cases} 0.50 & x=0 \\ 0.30 & x=1 \\ 0.15 & x=2 \\ 0.05 & x=3 \\ 0 & \text{otherwise} \end{cases}$$

### 4. Expectation $E[X]$
$$E[X] = \sum x \cdot P(X=x) = (0 \cdot 0.50) + (1 \cdot 0.30) + (2 \cdot 0.15) + (3 \cdot 0.05)$$
$$E[X] = 0 + 0.30 + 0.30 + 0.15 = 0.75$$

### 5. Variance $Var(X)$
First, find $E[X^2]$:
$$E[X^2] = (0^2 \cdot 0.50) + (1^2 \cdot 0.30) + (2^2 \cdot 0.15) + (3^2 \cdot 0.05) = 0.30 + 0.60 + 0.45 = 1.35$$
$$Var(X) = E[X^2] - (E[X])^2 = 1.35 - (0.75)^2 = 1.35 - 0.5625 = 0.7875$$

### 6. Probabilities
- $P(X \ge 1) = P(1) + P(2) + P(3) = 0.30 + 0.15 + 0.05 = 0.50$
- $P(X \ge 2) = P(2) + P(3) = 0.15 + 0.05 = 0.20$

### 7. Practical Meaning
- **Expectation:** The long-term average number of transformer failures per day.
- **Variance:** Measures the consistency of failures; a higher variance indicates a higher likelihood of days with zero or very high failure counts.

#Q4) Probability of Defective Items

### 1. Probability Space $(\Omega, \mathfrak{F}, P)$
- **Sample Space:** $\Omega = [0, 1]$
- **$\sigma$-algebra:** $\mathfrak{F} = \mathfrak{B}([0, 1])$ (Borel $\sigma$-algebra on the unit interval).
- **Probability Measure:** Defined by the integral of the Beta density function over sets in $\mathfrak{F}$.

### 3. Probability Density Function (PDF)
For $x \in (0, 1)$:
$$f(x) = \frac{1}{B(2, 18)} x^{2-1} (1-x)^{18-1} = \frac{1}{B(2, 18)} x (1-x)^{17}$$

### 4. Verify $f$ is valid
$f(x) \ge 0$ for all $x$. The integral over the support is:
$$\int_{0}^{1} \frac{x^{2-1}(1-x)^{18-1}}{B(2, 18)} dx = \frac{B(2, 18)}{B(2, 18)} = 1$$

### 5. Expectation $E[X]$
For a Beta distribution $\text{Beta}(\alpha, \beta)$:
$$E[X] = \frac{\alpha}{\alpha + \beta} = \frac{2}{2 + 18} = \frac{2}{20} = 0.10$$

### 6. Variance $Var(X)$
$$Var(X) = \frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)} = \frac{2 \cdot 18}{(20)^2(21)} = \frac{36}{400 \cdot 21} = \frac{36}{8400} \approx 0.00428$$

### 7. Probability $P(X \le 0.10)$
$$P(X \le 0.10) = \int_{0}^{0.10} \frac{1}{B(2, 18)} x (1-x)^{17} dx$$

### 8. Probability $P(0.05 \le X \le 0.15)$
$$P(0.05 \le X \le 0.15) = \int_{0.05}^{0.15} \frac{1}{B(2, 18)} x (1-x)^{17} dx$$

### 9. Why Beta is reasonable
The Beta distribution is ideal for modeling proportions or probabilities because its support is strictly bounded between $0$ and $1$, and it is highly flexible in shape depending on $\alpha$ and $\beta$.

### 10. Interpretation
- **Expectation (0.10):** The average expected defect rate in a production batch is $10\%$.
- **Variance:** Represents the stability of the production quality. A low variance means the defect rate is consistently around $10\%$.

# Q5) Expected Project Completion Time

### 1. Specify the support of $X$
The support is the set of all possible values for $X$. Since the shortest time is 4 hours and the longest is 10 hours:
$$Support(X) = [4, 10]$$

### 2. Write down the probability density function $f_X(x)$
For a Triangular distribution with $a=4, b=10, c=6$:
$$f_X(x) = \begin{cases} \frac{2(x-4)}{(10-4)(6-4)} = \frac{x-4}{6} & \text{for } 4 \le x < 6 \\ \frac{2(10-x)}{(10-4)(10-6)} = \frac{10-x}{12} & \text{for } 6 \le x \le 10 \\ 0 & \text{otherwise} \end{cases}$$

### 3. Verify that $f_X$ integrates to 1
The area of a triangle is $\frac{1}{2} \times \text{base} \times \text{height}$.
- Base = $b - a = 10 - 4 = 6$
- Height = $f_X(c) = f_X(6) = \frac{6-4}{6} = \frac{1}{3}$
$$Area = \frac{1}{2} \times 6 \times \frac{1}{3} = 1$$

### 4. Compute the expectation $E[X]$
$$E[X] = \frac{a+b+c}{3} = \frac{4+10+6}{3} = \frac{20}{3} \approx 6.67 \text{ hours}$$

### 5. Compute the variance $Var(X)$
$$Var(X) = \frac{a^2 + b^2 + c^2 - ab - ac - bc}{18} = \frac{16 + 100 + 36 - 40 - 24 - 60}{18} = \frac{28}{18} = \frac{14}{9} \approx 1.56$$

### 6. Interpretation of $a, b,$ and $c$
- **$a = 4$:** The "Optimistic" time; the absolute minimum time required if everything goes perfectly.
- **$b = 10$:** The "Pessimistic" time; the maximum time allowed even if delays occur.
- **$c = 6$:** The "Most Likely" time; the mode of the distribution where the probability density is highest.

### 7. Why Triangular distribution is reasonable
It is used when data is scarce but "expert judgment" can define the boundaries (best/worst case) and the peak. Unlike a Normal distribution, it has finite boundaries which are more realistic for physical tasks like installation.

### 8. Compute $P(X \le 5)$
Since $5$ is in the first interval $[4, 6)$:
$$P(X \le 5) = \int_{4}^{5} \frac{x-4}{6} dx = \left[ \frac{(x-4)^2}{12} \right]_{4}^{5} = \frac{(1)^2}{12} - 0 = \frac{1}{12} \approx 0.0833$$

### 9. Compute $P(5 \le X \le 8)$
$$P(5 \le X \le 8) = 1 - P(X < 5) - P(X > 8)$$
- $P(X < 5) = 1/12$
- $P(X > 8) = \int_{8}^{10} \frac{10-x}{12} dx = \left[ -\frac{(10-x)^2}{24} \right]_{8}^{10} = 0 - (-\frac{4}{24}) = \frac{1}{6}$
$$P(5 \le X \le 8) = 1 - \frac{1}{12} - \frac{2}{12} = \frac{9}{12} = 0.75$$

### 10. Use in Project Planning
This model helps project managers perform **Monte Carlo simulations** to estimate the probability of finishing a project by a certain deadline. It also allows for **sensitivity analysis** by changing $c$ (the most likely time) to see how it shifts the expected completion date and staffing needs.

# Q6) Expected Project Completion Time Using PERT

### 1. Compute the values of the shape parameters $\alpha$ and $\beta$
Given: $a = 8, b = 20, c = 11$.
$$\alpha = 1 + 4\left( \frac{c - a}{b - a} \right) = 1 + 4\left( \frac{11 - 8}{20 - 8} \right) = 1 + 4\left( \frac{3}{12} \right) = 1 + 1 = 2$$
$$\beta = 1 + 4\left( \frac{b - c}{b - a} \right) = 1 + 4\left( \frac{20 - 11}{20 - 8} \right) = 1 + 4\left( \frac{9}{12} \right) = 1 + 3 = 4$$

### 2. Support of $X$
The support is the interval of all possible completion times:
$$Support(X) = [8, 20]$$

### 3. Probability Density Function $f_X(x)$
Using $\alpha = 2, \beta = 4$ and the interval $[8, 20]$:
$$f_X(x) = \frac{(x - 8)^{2-1}(20 - x)^{4-1}}{B(2, 4)(20 - 8)^{2+4-1}} = \frac{(x - 8)(20 - x)^3}{B(2, 4) \cdot 12^5}, \quad 8 \le x \le 20$$

### 4. Probability Space $(\Omega, \mathfrak{F}, P)$
- **Sample Space:** $\Omega = [8, 20]$
- **$\sigma$-algebra:** $\mathfrak{F} = \mathfrak{B}([8, 20])$ (Borel sets on the interval).
- **Probability Measure:** $P(A) = \int_{A} f_X(x) dx$ for any $A \in \mathfrak{F}$.

### 5. Random Variable $X$
$X$ is the identity mapping $X(\omega) = \omega$ for $\omega \in [8, 20]$, representing the number of days required for task completion.

### 6. Compute the Expectation $E[X]$
$$E[X] = \frac{a + 4c + b}{6} = \frac{8 + 4(11) + 20}{6} = \frac{8 + 44 + 20}{6} = \frac{72}{6} = 12 \text{ days}$$

### 7. Compute the Variance $Var(X)$
Using the provided approximation:
$$Var(X) = \frac{(E[X] - a)(b - E[X])}{7} = \frac{(12 - 8)(20 - 12)}{7} = \frac{4 \times 8}{7} = \frac{32}{7} \approx 4.57$$

### 8. Approximate Standard Deviation $\sigma$
$$\sigma = \sqrt{Var(X)} = \sqrt{\frac{32}{7}} \approx 2.14 \text{ days}$$

### 9. PERT vs. Triangular Distribution
The PERT distribution is often more appropriate than the Triangular distribution because it is "smoother" and places less weight on the extreme values ($a$ and $b$). While the Triangular distribution has a sharp peak and linear slopes, the PERT distribution (based on the Beta distribution) produces a bell-like curve that more naturally models human estimation bias and uncertainty in software tasks.

### 10. Practical Interpretations
- **$a, b, c$:** Represent the best-case, worst-case, and most-likely scenarios provided by expert judgment.
- **$E[X] = 12$:** The weighted average completion time. Note that it is higher than the mode ($c=11$) because the distribution is skewed by the distant "worst-case" of 20 days.
- **$\sigma \approx 2.14$:** Represents the risk or uncertainty. A larger $\sigma$ indicates that the project manager should buffer more time to account for potential delays.

# Q7) If the PERT Expectation is to Match Our Intuition

### Problem Statement
A task duration $X$ is modeled by a PERT distribution with parameters $a$ (minimum), $b$ (maximum), and $c$ (most likely).  
Given:
- Minimum time $a = 5$ days
- Most likely time $c = 11$ days
- The expected time $E[X]$ is to be equal to the most likely time $c$.

### 1. Determine the value of the maximum time $b$
The formula for the expectation of a PERT distribution is:
$$E[X] = \frac{a + 4c + b}{6}$$

According to the problem, we want $E[X] = c$. We can set up the following equation:
$$\frac{a + 4c + b}{6} = c$$

Substitute the known values ($a = 5$ and $c = 11$):
$$\frac{5 + 4(11) + b}{6} = 11$$

Multiply both sides by 6:
$$5 + 44 + b = 66$$
$$49 + b = 66$$

Solve for $b$:
$$b = 66 - 49$$
$$b = 17$$

**The maximum time $b$ must be 17 days.**

### 2. What happens to the variance?
The variance of a PERT distribution is generally approximated by:
$$Var(X) = \frac{(E[X] - a)(b - E[X])}{7}$$

Substituting our values ($E[X] = 11, a = 5, b = 17$):
$$Var(X) = \frac{(11 - 5)(17 - 11)}{7}$$
$$Var(X) = \frac{6 \times 6}{7} = \frac{36}{7} \approx 5.14$$

**Analysis:**
When the expectation $E[X]$ is forced to equal the most likely value $c$, the distribution becomes **symmetric** around $c$ (since $c - a = 11 - 5 = 6$ and $b - c = 17 - 11 = 6$). In this symmetric state, the variance is maximized for that specific spread. If the distribution were skewed (if $b$ were much larger), the variance would increase further, but the expectation would no longer match the intuition of being the "most likely" value.
