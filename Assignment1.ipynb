# Q1) The Probability of Shooting at a Target

**Model Parameters:** Radius of disk $R = 10$.  
Event $A$: Point lies within 1 unit of the boundary $\implies 9 \le r \le 10$.

### 1. Uniform Distribution by Area
In this model, the probability is proportional to the area of the region.
- **Total Area:** $A_{total} = \pi \times 10^2 = 100\pi$
- **Area of inner disk ($r < 9$):** $A_{inner} = \pi \times 9^2 = 81\pi$
- **Area of Event $A$ (the outer ring):** $A_{A} = 100\pi - 81\pi = 19\pi$

$$P(A) = \frac{19\pi}{100\pi} = 0.19$$

### 2. Uniform Distribution by Distance and Direction
Here, $r$ is chosen uniformly from $[0, 10]$. The angle $\theta$ is independent and does not affect the radial probability.
- **Total range of $r$:** $10 - 0 = 10$
- **Favorable range for $A$:** $[9, 10]$, which has a length of $1$.

$$P(A) = \frac{10 - 9}{10} = \frac{1}{10} = 0.1$$

### 3. Explanation of Difference
The answers differ because the definition of "uniform" is applied to different variables. In **Model 1**, we assume every square unit of surface area is equally likely. In **Model 2**, we assume every increment of distance from the center is equally likely, which ignores the fact that there is physically more space (area) at the outer edges of a circle than at the center.


# Q2) Bertrand's Paradox

The side length of an inscribed equilateral triangle is $L = R\sqrt{3}$. We seek the probability that a random chord's length $X > R\sqrt{3}$.

### Method 1: Random Endpoints
1. Fix one point of the chord at a vertex of the triangle.
2. The chord is longer than $R\sqrt{3}$ only if the second point falls on the arc between the other two vertices.
3. This arc represents exactly $\frac{1}{3}$ of the total circumference.

$$P = \frac{1}{3}$$

### Method 2: Random Radial Distance
1. Choose a radius of the circle and pick a point on it uniformly to be the midpoint of the chord.
2. A chord is longer than the triangle side if its midpoint is closer to the center than $R/2$.
3. Since the point is chosen uniformly along the radius $R$:

$$P = \frac{R/2}{R} = \frac{1}{2}$$

### Method 3: Random Midpoint
1. Choose a point anywhere inside the circle uniformly to be the midpoint of the chord.
2. For the chord to be longer than $R\sqrt{3}$, the midpoint must fall within a smaller concentric circle of radius $r = R/2$.
3. The probability is the ratio of the area of the inner circle to the total area:

$$P = \frac{\pi (R/2)^2}{\pi R^2} = \frac{\frac{1}{4}\pi R^2}{\pi R^2} = \frac{1}{4}$$

# Q8) Probability of Being a Girl?

### Problem Statement
A family has two children. Each child is equally likely to be a boy ($B$) or a girl ($G$), independently of the other. We want to find the probability that both children are girls, given that at least one of them is a girl.

### 1. Define the Sample Space
The sample space $\Omega$ for two children (where order matters, e.g., first child, second child) is:
$$\Omega = \{ (B, B), (B, G), (G, B), (G, G) \}$$
Since each outcome is equally likely, $P(\omega) = \frac{1}{4}$ for each $\omega \in \Omega$.

### 2. Define the Events
* Let event $A$ be "both children are girls":
    $$A = \{ (G, G) \}$$
* Let event $B$ be "at least one child is a girl":
    $$B = \{ (B, G), (G, B), (G, G) \}$$

### 3. Apply Conditional Probability
We are looking for $P(A | B)$, which is defined as:
$$P(A | B) = \frac{P(A \cap B)}{P(B)}$$

* **Step 1: Find $P(A \cap B)$** The intersection $A \cap B$ represents outcomes where both children are girls AND at least one is a girl. This is simply the set $\{(G, G)\}$.
    $$P(A \cap B) = \frac{1}{4}$$

* **Step 2: Find $P(B)$** Event $B$ contains 3 out of the 4 possible outcomes.
    $$P(B) = \frac{3}{4}$$

* **Step 3: Calculate the Final Probability** $$P(A | B) = \frac{1/4}{3/4} = \frac{1}{3}$$

### Final Answer
The probability that both children are girls, given that at least one is a girl, is:
$$\mathbf{P = \frac{1}{3}}$$

# Q9) Information Available in a Discrete Random Variable

### 1. Probability Space $(\Omega, \mathfrak{F}, P)$
The probability space consists of the sample space $\Omega$, the $\sigma$-algebra $\mathfrak{F}$, and the probability measure $P$.
- **Sample Space:** $\Omega = \{G_1, G_2, D_1, D_2\}$
- **$\sigma$-algebra:** $\mathfrak{F} = \mathfrak{P}(\Omega)$ (The power set, containing $2^4 = 16$ events).
- **Probability Measure:** - $P(\{G_1\}) = 0.50$
  - $P(\{G_2\}) = 0.20$
  - $P(\{D_1\}) = 0.10$
  - $P(\{D_2\}) = 0.20$

### 2. $\sigma$-algebra generated by $X$
$X$ partitions $\Omega$ into two sets: $\{G_1, G_2\}$ (where $X=0$) and $\{D_1, D_2\}$ (where $X=1$).
$$\sigma(X) = \{ \emptyset, \{G_1, G_2\}, \{D_1, D_2\}, \Omega \}$$

### 3. $\sigma$-algebra generated by $Y$
Since $Y$ assigns a unique value to every outcome, it generates the power set.
$$\sigma(Y) = \{ \emptyset, \{G_1\}, \{G_2\}, \{D_1\}, \{D_2\}, \{G_1, G_2\}, \dots, \Omega \} = \mathfrak{P}(\Omega)$$

### 4. Show that $\sigma(X) \subseteq \sigma(Y)$
Every set in $\sigma(X)$ is also a set in $\sigma(Y)$ (e.g., $\{G_1, G_2\} = \{G_1\} \cup \{G_2\}$).
**Explanation:** $Y$ contains more information than $X$ because knowing $Y$ tells us both the product quality **and** the specific machine that produced it, whereas $X$ only identifies the quality.

### 5. Marginal Distribution of $X$
- $P(X=0) = P(\{G_1, G_2\}) = 0.50 + 0.20 = 0.70$
- $P(X=1) = P(\{D_1, D_2\}) = 0.10 + 0.20 = 0.30$

### 6. Marginal Distribution of $Y$
- $P(Y=1) = 0.50, \quad P(Y=2) = 0.20, \quad P(Y=3) = 0.10, \quad P(Y=4) = 0.20$

### 7. Joint Probabilities
- $P(X=0, Y=1) = P(\{G_1\}) = 0.50$
- $P(X=0, Y=2) = P(\{G_2\}) = 0.20$
- $P(X=1, Y=3) = P(\{D_1\}) = 0.10$
- $P(X=1, Y=4) = P(\{D_2\}) = 0.20$

### 8. Conditional Probabilities
- $P(X=1 | Y=3) = \frac{P(X=1, Y=3)}{P(Y=3)} = \frac{0.10}{0.10} = 1$
- $P(X=1 | Y=4) = \frac{0.20}{0.20} = 1$
- $P(Y=3 | X=1) = \frac{P(X=1, Y=3)}{P(X=1)} = \frac{0.10}{0.30} = \frac{1}{3}$
- $P(Y=4 | X=1) = \frac{P(X=1, Y=4)}{P(X=1)} = \frac{0.20}{0.30} = \frac{2}{3}$

### 9. Difference in Conditioning
Conditioning on $Y$ provides complete certainty about $X$ because $Y$ is a "finer" variable. Conditioning on $X$ only narrows down the possibilities for $Y$ to a subset (the quality type), but leaves uncertainty regarding which machine produced it.

### 10. Information Comparison
- **In $\sigma(Y)$ but not in $\sigma(X)$:** Information about the specific machine (Machine 1 vs Machine 2).
- **In $\sigma(X)$ but not in $\sigma(Y)$:** None. Since $\sigma(X) \subseteq \sigma(Y)$, all information in $X$ is inherently available in $Y$.

### 11. Python Visualization Logic
To visualize this, one could use a bar chart showing the probabilities of $\Omega$ or a heat map of the joint distribution $P(X, Y)$.

### 12. Shannon Entropy (in bits)
- **$H(X)$:** $-(0.7 \log_2 0.7 + 0.3 \log_2 0.3) \approx 0.881$
- **$H(Y)$:** $-(0.5 \log_2 0.5 + 0.2 \log_2 0.2 + 0.1 \log_2 0.1 + 0.2 \log_2 0.2) \approx 1.761$
- **$H(X, Y)$:** Since $Y$ determines $X$, $H(X, Y) = H(Y) \approx 1.761$
- **$H(X|Y)$:** $0$ (knowledge of $Y$ removes all uncertainty of $X$).
- **$H(Y|X)$:** $H(Y) - H(X) \approx 0.880$

# Q10) Information Available in Continuous Random Variables

### 1. Probability Space $(\Omega, \mathfrak{F}, P)$
- **Sample Space:** $\Omega = \mathbb{R}^2$
- **$\sigma$-algebra:** $\mathfrak{F} = \mathfrak{B}(\mathbb{R}^2)$ (the Borel $\sigma$-algebra on $\mathbb{R}^2$)
- **Probability Measure:** $P$ is the measure induced by the joint density $f_{T,P}(t,p) = \frac{1}{4\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right)$.

### 2. $\sigma$-algebra generated by $X$, $\sigma(X)$
$X$ is the projection onto the first coordinate ($T$). Therefore, $\sigma(X)$ consists of all Borel sets in $\mathbb{R}^2$ that are cylinders over the $T$-axis.
$$\sigma(X) = \{ B \times \mathbb{R} : B \in \mathfrak{B}(\mathbb{R}) \}$$

### 3. $\sigma$-algebra generated by $Y$, $\sigma(Y)$
Since $Y(t,p) = (t,p)$ is the identity map on $\mathbb{R}^2$, it generates the entire Borel $\sigma$-algebra of the space.
$$\sigma(Y) = \mathfrak{B}(\mathbb{R}^2)$$

### 4. Show $\sigma(X) \subseteq \sigma(Y)$
Since every set of the form $B \times \mathbb{R}$ is a Borel set in $\mathbb{R}^2$, it follows that $\sigma(X) \subseteq \sigma(Y)$.  
**Reasoning:** $Y$ contains more information than $X$ because $Y$ records the state of both sensors (Temperature and Pressure), whereas $X$ only records Temperature. Knowing $Y$ allows you to determine $X$ perfectly, but knowing $X$ leaves the value of $P$ unknown.

### 5. Marginal Density of $X$
$X = T$, and $T \sim N(0,1)$.
$$f_X(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}$$

### 6. Marginal Density of $Y$
$Y$ is the random vector $(T, P)$. Its density is the joint density provided:
$$f_Y(t,p) = \frac{1}{4\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right)$$

### 7. Probabilities
- $P(X \le 0) = \int_{-\infty}^{0} f_X(x) dx = 0.5$
- $P(X > 1) = 1 - \Phi(1) \approx 1 - 0.8413 = 0.1587$
- $P(Y \in (-\infty, 0] \times \mathbb{R}) = P(T \le 0) = 0.5$
- $P(Y \in [-1, 1] \times [-2, 2]) = P(-1 \le T \le 1) \times P(-2 \le P \le 2)$
  - $P(-1 \le T \le 1) \approx 0.6827$
  - $P(-2 \le P \le 2)$ where $P \sim N(0, 4) \implies Z = P/2 \sim N(0,1) \implies P(-1 \le Z \le 1) \approx 0.6827$
  - Result: $0.6827 \times 0.6827 \approx 0.4661$

### 8 & 9. Conditional Distributions
- **$X$ given $Y=(t,p)$:** Since $Y$ includes $X$, the distribution is a Dirac delta (point mass) at $X=t$.
- **$Y$ given $X=t$:** Since $T$ and $P$ are independent, $Y|X=t$ is effectively the distribution of $(t, P)$ where $P \sim N(0,4)$.

### 10. Interpretation
Conditioning on $X$ fixes the temperature but leaves the pressure $P$ unknown (random). Conditioning on $Y$ fixes both variables, leaving no randomness in the system.

### 11. Measurability
Since $X(t,p) = t = \pi_1(t,p)$, and $\pi_1$ is a continuous projection map, it is Borel measurable. This implies $X$ is a measurable function of $Y$.

### 13 & 14. Differential Entropy
- $h(X) = \frac{1}{2} \ln(2\pi e \sigma_T^2) = \frac{1}{2} \ln(2\pi e) \approx 1.4189 \text{ nats}$
- $h(Y) = h(T) + h(P) = \frac{1}{2} \ln(2\pi e) + \frac{1}{2} \ln(2\pi e \cdot 4) = \ln(4\pi e) \approx 3.4838 \text{ nats}$

### 15. Conditional Differential Entropy $h(Y|X)$
$h(Y|X) = h(T,P|T) = h(P|T)$. Because they are independent:
$$h(Y|X) = h(P) = \frac{1}{2} \ln(8\pi e) \approx 2.0649 \text{ nats}$$
**Physical Interpretation:** This represents the average uncertainty remaining in the full sensor state after the temperature has been measured.

### 16. The case of $h(X|Y)$
In this model, $h(X|Y) = -\infty$.  
**Explanation:** Unlike discrete variables where $H(X|Y)=0$ when $Y$ determines $X$, in the continuous case, if $X$ is a function of $Y$, the "uncertainty" collapses to a point with zero width, leading to negative infinite differential entropy.
